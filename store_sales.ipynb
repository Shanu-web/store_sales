#### Top 10 product by Sales

```python
product_sales=store_data.groupby("product_name").sum().sales.reset_index()
```

```python
top_prod_sales=product_sales.sort_values("sales", ascending=False).head(10)
```

```python
top_prod_sales
```

```python
plt.figure(figsize=(15,6))
font1 = {'family':'serif','color':'black','size':20}
plt.title("Top 10 product", fontdict=font1)
plt.bar(top_prod_sales["product_name"],top_prod_sales["sales"])
plt.xticks(rotation='vertical', size=12)
plt.show()
```

#### Most Selling Product
